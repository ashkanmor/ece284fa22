```python
### Training with manually updating W with "Backward" ###

import torch
#from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim




data = [(1.0,2.1,3.0), (2.0, 3.5, 6.0), (3.0, 3.0, 9.0), (4.0, 2.1, 12.0), (5.0, 7.2, 15.0), (6.0, 10.1, 18.0)]


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 1, bias=False)
    def forward(self, x):
        x = self.fc1(x)
        return x
        

net = Net()

print(net)
print(list(net.parameters()))

#input = torch.randn(1)
#out = net(input)

#def criterion(out, label):
#    return (label - out)**2
criterion = nn.MSELoss()


optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
#optimizer = optim.Adam(net.parameters(), lr=0.005)

minibatch_size = 3
data_tensor = torch.split(torch.FloatTensor(data), 3)

for epoch in range(30):
    for i in range (len(data_tensor)):
        X = []
        Y = []
        for j, current_data in enumerate(data_tensor[i]):
            X_temp1, X_temp2, Y_temp = current_data
            X_temp = [X_temp1, X_temp2]
            X.append(X_temp)
            Y.append([Y_temp])
        X = torch.FloatTensor(X)
        Y = torch.FloatTensor(Y)
        optimizer.zero_grad()
        outputs = net(X)
        loss = criterion(outputs, Y)
        loss.backward()
        optimizer.step()
        print("Epoch {} - MiniBatch {} - Loss {}".format(epoch,i+1,loss))
        
        
#test the network

for i, current_data in enumerate(data):
    X1, X2, Y = current_data
    X = [X1, X2]
    X = torch.FloatTensor(X)
    out = net(X)
    print("When X = {}, Y = {}".format(X, out))
```
